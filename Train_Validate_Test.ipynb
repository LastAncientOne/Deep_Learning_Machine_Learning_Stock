# Training set, Testing set, and Validation sets


Training set: data set that will be use for algorithm and fit the model. 

Validation set: data set that be use for unbiased evaluation of a model fit on the training dataset while tuning model hyperparameters. It involves randomly dividing the available set of observations into two parts, a training set and a validation set or hold-out set. The model is fit on the training set, and the fitted model is used to predict the responses for the observations in the validation set.

Testing set: data set that will be use for unbiased evaluation of a final model fit on the training set.        


Validation set is to find the values for the hyperparmeters and use it to evaluate the performance of the model for different combination of hyperparameter values.

```python
def train_validate_test_split(df, train_percent=.6, validate_percent=.2):
    perm = np.random.permutation(df.index)
    m = len(df.index)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    train = df.ix[perm[:train_end]]
    validate = df.ix[perm[train_end:validate_end]]
    test = df.ix[perm[validate_end:]]
    return train, validate, test
```

```python
# Leave one out cross validation (LOOCV)
# Lower bias, but higher variation 
from sklearn.model_selection import LeaveOneOut
X = np.array(dataset[['High', 'Low']])
y = np.array(dataset['Adj Close'])
loo = LeaveOneOut()
loo.get_n_splits(X)

for train_index, test_index in loo.split(X):
        print("train:", train_index, "validation:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
```

```python
# Lower K value is more bised
# Higher K value is less biased, but can suffer from large variability. 
# Smaller K value always takes us towards validation set approach, whereas a higher value of k leads to LOOCV approach.

from sklearn.model_selection import RepeatedKFold
kf = RepeatedKFold(n_splits=5, n_repeats=10, random_state=None) 

for train_index, test_index in kf.split(X):
      print("Train:", train_index, "Validation:",test_index)
      X_train, X_test = X[train_index], X[test_index] 
      y_train, y_test = y[train_index], y[test_index]
```

```python
# Stratified k-fold cross validation
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, random_state=None)

y = dataset['Buy_Sell']

# X is the feature set and y is the target
# Target must be 'binary' or 'multiclass'
for train_index, test_index in skf.split(X,y): 
    print("Train:", train_index, "Validation:", test_index) 
    X_train, X_test = X[train_index], X[test_index] 
    y_train, y_test = y[train_index], y[test_index]
```

```python
# k-fold cross validation with repetition
# Repeated cross-validation, the cross-validation procedure is repeated n times, yielding n random partitions of the original sample. 
# The n results are again averaged (or otherwise combined) to produce a single estimation.
from sklearn.model_selection import RepeatedKFold
rkf = RepeatedKFold(n_splits=5, n_repeats=10, random_state=None)

# X is the feature set and y is the target
for train_index, test_index in rkf.split(X):
     print("Train:", train_index, "Validation:", test_index)
     X_train, X_test = X[train_index], X[test_index]
     y_train, y_test = y[train_index], y[test_index]
```

```python
# Cross Validation for time series
# Splitting a time-series dataset randomly and is used for forecasting problem
from sklearn.model_selection import TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits=3)

for train_index, test_index in tscv.split(X):
     print("Train:", train_index, "Validation:", test_index)
     X_train, X_test = X[train_index], X[test_index]
     y_train, y_test = y[train_index], y[test_index]
```
