# PyCaret Stock Prediction

```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

# yahoo finance used to fetch data 
import yfinance as yf
yf.pdr_override()
```

```python
df= yf.download("AMD", start="2020-01-01", end="2022-01-01")
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
df = df.drop('Date', axis=1)
```

```python
data = df.sample(frac=0.9)
data_unseen = df.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.regression import *
exp_reg102 = setup(data=data, target='Adj Close', session_id=123,use_gpu=True)
```

```python
compare_models()
```

```python
lr = create_model('lr')
```

```python
tuned_lr = tune_model(lr, n_iter=1000)
```

```python
unseen_predictions = predict_model(tuned_lr, data=data_unseen)
unseen_predictions
```
